```python
#Exercicio 4
import unicodedata

# Passo 1: Receber o número de produtos eletrônicos do usuário
num_entradas = int(input("Digite o número de produtos eletrônicos: "))
produtos = []

# Passo 2: Receber nomes e preços na mesma linha
for i in range(num_entradas):
    entrada = input(f"Digite o nome e o preço do produto {i+1} (separados por espaço): ")
    partes = entrada.rsplit(' ', 1)  # Dividir a entrada em duas partes: nome e preço

    # Verificar se a entrada possui o formato correto (nome e preço separados por espaço)
    if len(partes) != 2:
        print("Formato incorreto. Digite o nome e o preço separados por espaço.")
        continue

    nome = partes[0]
    preco = float(partes[1])

    # Substituir caracteres especiais
    nome = unicodedata.normalize('NFKD', nome).encode('ASCII', 'ignore').decode('ASCII')

    produtos.extend([nome, preco])

# Passo 3: Criar listas separadas para nomes e preços
nomes = produtos[::2]
precos = produtos[1::2]

# Passo 4: Ordenar a lista de nomes
nomes_ordenados = sorted(nomes)

# Passo 5: Imprimir os nomes ordenados e seus respectivos preços
for nome, preco in zip(nomes_ordenados, precos):
    print(f"{nome}: R${preco}")

# Criar a lista final
lista_final = produtos

# Imprimir a lista final
print(lista_final)

```

```python
#Exercicio 7
import re;
import random;

nomes = ["Luana", "Ju", "Davi", "Vivi", "Pri", "Cleo"]

def length(lista):
    n = int(input('Digite um número n'))
    maior = 0
    for item in lista:
        maior += 1 if len(item) > n else 0

    return maior

def vogais(lista):
    numVogais = []

    for item in lista:
        resultado = re.findall('[aA,Ee,Ii,Oo,Uu]{1}', item)
        numVogais.append(len(resultado))

    return numVogais

def group_by_legth(lista):
    retorno = []
    lista.sort(key = lambda x : len(x))

    listaAux = lista.copy()

    while len(listaAux) > 0:
        conjunto = list(filter(lambda item : len(listaAux[0]) == len(item), listaAux))
        retorno.append(conjunto)

        for item in conjunto:
            listaAux.remove(item)

    return retorno

def encrypt(lista):
    key = random.randint(1, 10)
    retorno = []

    for indexItem, item in enumerate(lista):
      string = ''
      for indexLetra, letra in enumerate(item):
        valorLetra = ord(letra)
        valorLetra += key
        string += chr(valorLetra)
      retorno.append(string)

    return f"{retorno}, {key}"

encerrar = False

while not encerrar:
    print('Selecione uma das opções abaixo:')
    print('[1] - Length')
    print('[2] - Vogais')
    print('[3] - Group By')
    print('[4] - Encrypt')
    print('[0] - Encerrar o programa')
    escolha = input()

    if escolha == '1':
        print(length(nomes))
    elif escolha == '2':
        print(vogais(nomes))
    elif escolha == '3':
        print(group_by_legth(nomes))
    elif escolha == '4':
        print(encrypt(nomes))
    elif escolha == '0':
        encerrar = True
    else:
        print('Escolha inválida')
```
